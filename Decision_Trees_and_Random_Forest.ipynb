```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import pandas as pd
```

```python
wine = load_wine()

X = pd.DataFrame(wine.data, columns=[wine.feature_names])
Y = pd.Series(wine.target, name='wine_class')

print(X.head())

print(Y.head())

# Let's see what the different wine classes are
print("\nOur target classes are:", wine.target_names)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)

print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
```

```python
dt_model = DecisionTreeClassifier(random_state=42)

dt_model.fit(X_train, Y_train)
```

```python
dt_predict = dt_model.predict(X_test)

dt_accuracy = accuracy_score(Y_test, dt_predict)

print(f"Decision Tree Accuracy: {dt_accuracy * 100:.2f}%")
```

**Now Random Forest**

```python
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

rf_model.fit(X_train, Y_train)
```

```python
rf_predict = rf_model.predict(X_test)

rf_accuracy = accuracy_score(Y_test, rf_predict)

print(f"Random Forest Accuracy: {rf_accuracy * 100:.2f}%")
```

```python
# --- Final Comparison ---

print("\n--- Model Performance ---")
print(f"Decision Tree Accuracy: {dt_accuracy * 100:.2f}%")
print(f"Random Forest Accuracy: {rf_accuracy * 100:.2f}%")
print("-----------------------")

if rf_accuracy > dt_accuracy:
    print("\nThe Random Forest model performed better!")
else:
    print("\nThe Decision Tree model performed better or the same!")
```
